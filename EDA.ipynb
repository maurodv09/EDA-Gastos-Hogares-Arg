Comparacion entre número de habitantes y número de encuestas:

```python
#mapa poblacion provincias
prov_pop = {
    'CABA':2890151,
    'Buenos Aires':15625084,
    'Catamarca':367828,
    'Córdoba':3308876,
    'Corrientes':992595,
    'Chaco':1055259,
    'Chubut':509108,
    'Entre Ríos':1235994,
    'Formosa':530162,
    'Jujuy':673307,
    'La Pampa':318951,
    'La Rioja':333642,
    'Mendoza':1738929,
    'Misiones':1101593,
    'Neuquén':551266,
    'Río Negro':638645,
    'Salta':1214441,
    'San Juan':681055,
    'San Luís':432310,
    'Santa Cruz':273964,
    'Santa Fe':3194537,
    'Santiago del Estero':874006,
    'Tierra del Fuego':127205,
    'Tucumán':1448188
}
#source: https://www.ign.gob.ar/NuestrasActividades/Geografia/DatosArgentina/Poblacion2
```

```python
prov_df = (                         #crea df con la poblacion y el numero de encuentas normalizadas
    pd.DataFrame(prov_pop.items(), columns=['prov_name', 'poblacion'])
        .join(df.prov_name.value_counts(normalize=True), on='prov_name', how='outer')
        .set_index('prov_name') )
prov_df.poblacion = prov_df.poblacion / prov_df.poblacion.sum() #normaliza valores de poblacion
prov_df.rename(columns={'proportion': 'encuestas'}, inplace=True) 
prov_df.sort_values('poblacion', ascending=False, inplace=True) #ordena por poblacion
```

```python
prov_df.plot(kind='bar', title='Población vs Encuestas (normalizado)', xlabel='', figsize=(8,4))
plt.show()
```

### Miembros (Tipo de hogares)

#### Composición de los Hogares

```python
fig, axs = plt.subplots(1,3, figsize=(12,4))
axs[0].scatter(df.cantmiem, df.cantping, alpha=.05)
axs[0].set(ylim=(-1,25), xlim=(-1,25), xlabel='cantmiem', ylabel='cantping', title='Miembros con ingresos vs Miembros Totales')

axs[1].scatter(df.cantmiem, df.reldep, alpha=.05)
axs[1].set(xlabel='cantmiem', ylabel='reldep', title='Rel. dependencia vs Miembros Totales')

axs[2].scatter(df.cantping, df.reldep, alpha=.2)
axs[2].set(xlabel='cantping', ylabel='reldep', title='Rel. dependencia vs Miembros con Ingresos')

plt.tight_layout()
```

A primera vista podemos notar:
- De entre los 1 a 7-8 miembros que componen la gran mayoria de los hogares argentinos, la cantidad de miembros con ingresos parece crecer linealmente.
- La relación de dependencia también crece conforme al tamaño de los hogares, como es esperable.
- Se observa una gran variabilidad en la relación de dependencia para hogares con pocas personas con percepción de ingresos.

#### Gastos e Ingresos por Hogar

```python
fig, axs = plt.subplots(2,2, figsize=(12,5))
fig.suptitle('Gastos e Ingresos por número de Miembros en el hogar')

sns.scatterplot(df, x='cantmiem', y='ingtoth', alpha=.5, ax=axs[0,0])
axs[0,0].set(ylabel='Ingreso Total', xlabel='Número de Miembros')

sns.scatterplot(df, x='cantmiem', y='gastot', color='red', alpha=.5, ax=axs[0,1])
axs[0,1].set(ylabel='Gastos Totales', xlabel='Número de Miembros')

sns.scatterplot(df, x='cantmiem', y='ingpch', alpha=.5, ax=axs[1,0])
axs[1,0].set(ylabel='Ingresos per capita', xlabel='Número de Miembros')

sns.scatterplot(df, x='cantmiem', y='gastotpc', color='red', alpha=.5, ax=axs[1,1])
axs[1,1].set(ylabel='Gastos per capita', xlabel='Número de Miembros')

plt.tight_layout()
```

- Notar la pobreza relativa (ingresos y gastos totales) para hogares altamente poblados

##### Escala Logarítmica

```python
for col in ['ingtoth', 'ingpch', 'gastot', 'gastotpc']: 
    df['log_' + col] = df[col].apply(lambda x: np.log(x+1)) #transformacion logaritmica
```

```python
fig, axs = plt.subplots(2,2, figsize=(12,6))
fig.suptitle('Gastos e Ingresos por N° de Miembros (Log scale regression)')
sns.regplot(df, x='cantmiem', y='log_ingtoth', color='red', scatter_kws={'alpha': 0.05, 'color':'blue'}, ax=axs[0,0])
axs[0,0].set(ylabel='Ingresos Totales', xlabel='Número de Miembros', ylim=(-.5,15.5))

sns.regplot(df, x='cantmiem', y='log_gastot', color='red', scatter_kws={'alpha': 0.05, 'color':'blue'}, ax=axs[0,1])
axs[0,1].set(ylabel='Gastos Totales', xlabel='Número de Miembros', ylim=(-.5,15.5))

sns.regplot(df, x='cantmiem', y='log_ingpch', color='red', scatter_kws={'alpha': 0.05, 'color':'blue'}, ax=axs[1,0])
axs[1,0].set(ylabel='Ingresos per capita', xlabel='Número de Miembros', ylim=(-.5,15.5))

sns.regplot(df, x='cantmiem', y='log_gastotpc', color='red', scatter_kws={'alpha': 0.05, 'color':'blue'}, ax=axs[1,1])
axs[1,1].set(ylabel='Gastos per capita', xlabel='Número de Miembros', ylim=(-.5,15.5))

plt.tight_layout()
```

- Como visto en el mapa de correlación, el ingreso total y el gasto total no incrementa en mayor medida conforme aumenta el número de miembros, a pesar de la alta correlación entre el total de miembros y el número de ellos que perciben ingresos. Esto explica en cierta medida la notable disminución de ingresos y gastos per capita en los hogares.
- Notar también que, naturalmente, estos patrones no encajan correctamente para hogares excesivamente poblados, por lo que sería necesario un análisis en profundidad solo para estos casos.
