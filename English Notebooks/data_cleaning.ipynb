The dataset is composed of 134 columns, from which will use:

    * gc_05 Gasto de consumo división 5 - Home equipment and maintenance
    * gc_06 Gasto de consumo división 6 - Health services
    * gc_07 Gasto de consumo división 7 - Transportation
    * gc_08 Gasto de consumo división 8 - Comunications
    * gc_09 Gasto de consumo división 9 - Recreation and culture
    * gc_10 Gasto de consumo división 10 - Education
    * gc_11 Gasto de consumo división 11 - Restaurants y hotels
    * gc_12 Gasto de consumo división 12 - Other goods y services

4. formas de pago
    * fp_contado Total expenses in cash
    * fp_credito Total expenses in credits
    * fp_especie Total expenses in kind

##### Calculations Consistency

```python
hog['gastot_calculado'] = hog[['gc_01', 'gc_02', 'gc_03', 'gc_04', 'gc_05', 'gc_06', 'gc_07', 'gc_08', 'gc_09', 'gc_10', 'gc_11', 'gc_12']].sum(axis=1)
print('any error total expenses calculation?', (abs(hog.gastot - hog.gastot_calculado) > 0.1).any())
```

```python
hog.drop('gastot_calculado', axis=1, inplace=True)
```

##### gastotpc:

```python
print('any gastotpc !=  gastot / cantmiem?', (abs(hog.gastotpc - hog.gastot / hog.cantmiem) > 0.1).any())
```

#### Payment Method

```python
hog['fp_diff'] = hog['gastot'] - hog[['fp_contado', 'fp_credito', 'fp_especie', 'fp_prodprop', 'fp_tarjetas', 'fp_otras']].sum(axis=1)
hog['fp_diff_perc'] = hog['fp_diff'] / hog['gastot'] * 100
```

```python
print('Toleretad percentege error - percentage of number of errors in the dataset')
for i in range(1,16,2):
    print(f'{i:02d}%', '- ', f'{round((abs(hog.fp_diff_perc) > i).sum()/hog.shape[0] * 100,2)}%')
```

```python
fp_er = hog.loc[abs(hog.fp_diff) > 1, 'fp_diff'] / hog.loc[abs(hog.fp_diff) > 1, 'gastot'] * 100
```

```python
fp_er.hist(bins=20, figsize=(6,2))
plt.title('Percentage Error distribution in fp variables')
plt.show()
```

```python
fp_er.loc[lambda x: x >= 1].apply(lambda x: np.log(x)).hist(bins=20, figsize=(6,2))
plt.title('Porcentage Error log distribution in fp variables (positives)')
plt.show()
```

```python
fp_er.loc[lambda x: x <= -1].apply(lambda x: np.log(-x)).hist(bins=20, figsize=(6,2))
plt.title('Porcentage Error log distribution in fp variables (negatives)')
plt.show()
```

The number and the variaty of inconsistencies is considerable large, so we'll ultimately won't use these characteristics for the rest of the analysis

```python
hog.drop(columns=['fp_contado', 'fp_credito', 'fp_especie', 'fp_prodprop', 'fp_tarjetas','fp_otras', 'fp_indef', 'fp_diff', 'fp_diff_perc'], inplace=True)
```
